```python
age = 2022 - df['Date_of_birth']
df['age'] = age
```

```python
# Filtering out the data to show the candidates between ages 16 to 35
df = df[df['age'].between(16, 35)]
```

```python
df
```

## Inclusion Criteria

```python
df.columns
```

```python

# Person who are not on on anticoagulant therapy
df = df[df.Anticoagulant_therapy == "no"]

# Person who are not on contraceptive pill, pregnant or breastfeeding
df = df[df.Contraceptive_Pill_Pregnant_or_Breastfeeding == "no"]

# Person who are not taking any treatment
df = df[df.Treatment_Acne == "no"]

# Person who are having persistent acne
df = df[df.Do_you_have_persistent_acne_grade_II_and_III == "yes"]

# Person who uses anitmicrobial washes
df = df[df.Using_Standard_Antimicrobial_Washes == "yes"]
```

```python
# Calculating the percentage
total_number_per = (df.First_Name.count()/2499)*100
print("Estimated percentage of the population that might be eligible for, and able to participate in a clinical trial with the below eligibility criteria and physical location." ,round(total_number_per),"%")
```
